```python
import cv2
import numpy as np
import mediapipe as mp
import time
import os
import pandas as pd

drawing_utils = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
train_dir = "data2/"

counter = 0
trash = 0
trashFiles = []

data = pd.DataFrame()
import os
for _, dirs, _ in os.walk(train_dir):
    for dir in dirs:
        for dirname, _, files in os.walk(train_dir+dir):
            for file in files:
                counter += 1
                if (counter % 100 == 0):
                    print(str(counter) + os.path.join(dirname, file))

                image = cv2.imread(os.path.join(dirname, file))
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                with mp_holistic.Holistic(min_detection_confidence=0.4, min_tracking_confidence=0.4) as holistic:
                    results = holistic.process(image)

                    # drawing_utils.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                    #                           drawing_utils.DrawingSpec(color=(80, 110, 10), thickness=1, circle_radius=1),
                    #                           drawing_utils.DrawingSpec(color=(80, 256, 121), thickness=1, circle_radius=1))

                    # drawing_utils.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
                    # drawing_utils.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
                    # drawing_utils.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)


                    if not results.face_landmarks:
                        trash+=1
                        trashFiles.append(os.path.join(dirname, file))
                        print("err: " + os.path.join(dirname, file))
                        continue

                    i = 0
                    vals = []
                    for landmark in results.face_landmarks.landmark:
                        vals.append(landmark.x)
                        vals.append(landmark.y)
                        vals.append(landmark.z)

                    data = pd.concat([data, pd.DataFrame([vals])], ignore_index=True)

                # if cv2.waitKey(1) == 27:  # exit on ESC
                #     break
```
