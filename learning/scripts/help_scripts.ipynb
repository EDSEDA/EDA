Скрипт для разбиения изображений на тренировочную и тестовую выборку

```python
import os
import random
import shutil

# Путь к директории 'train'
train_dir = 'data_large/train'
# Путь к директории 'test'
test_dir = 'data_large/test'

# Список эмоций
emotions = ['anger', 'disgust', 'fear', 'happy', 'neutral', 'sadness', 'surprise']

# Проход по каждой эмоции
for emotion in emotions:
    # Создайте поддиректорию в 'test', если её нет
    test_emotion_dir = os.path.join(test_dir, emotion)
    os.makedirs(test_emotion_dir, exist_ok=True)

    # Получите список файлов в директории 'emotion'
    emotion_dir = os.path.join(train_dir, emotion)
    image_files = os.listdir(emotion_dir)

    # Выберите случайные 10% файлов
    num_files_to_move = int(0.1 * len(image_files))
    files_to_move = random.sample(image_files, num_files_to_move)

    # Переместите выбранные файлы в 'test/emotion'
    for file in files_to_move:
        src = os.path.join(emotion_dir, file)
        dst = os.path.join(test_emotion_dir, file)
        shutil.move(src, dst)

print("Готово! Изображения перенесены в директорию 'test'.")
```

Скрипт для конветации фрмата изображений

```python
from PIL import Image
import os

directory = 'data_large/train'

# Список эмоций
emotions = ['anger', 'disgust', 'fear', 'happy', 'neutral', 'sadness', 'surprise']

# Проход по каждой эмоции
for emotion in emotions:
    emotion_dir = os.path.join(directory, emotion)
    for filename in os.listdir(emotion_dir):
        print(filename)
        if filename.endswith(".png"):
            with Image.open(os.path.join(emotion_dir, filename)) as img:
                output_filename = os.path.splitext(filename)[0] + ".jpg"
                img.save(os.path.join(emotion_dir, output_filename), "JPEG")
                os.remove(os.path.join(emotion_dir, filename))
                
    
print("Конвертация завершена.")
```
